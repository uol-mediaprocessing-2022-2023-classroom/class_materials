# Digital Images


```
import numpy as np
import cv2
from skimage import io
from matplotlib import pyplot as plt
```

First, we need to load an example image. Since **OpenCV** cannot download images natively, we use the **wget** command to download it to the server. Then, we need to load and convert it to the **rgb colorspace** with the **cvtColor** function.

```
! wget -q https://raw.githubusercontent.com/svenschultze/mediaprocessing-notebooks/master/pictures/Lenna.png
img = cv2.imread("Lenna.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

Alternatively, with the library **skimage** (scikit-image), we can load images directly from the internet in the rgb format:

```
img = io.imread('https://raw.githubusercontent.com/svenschultze/mediaprocessing-notebooks/master/pictures/Lenna.png')
```

Now, we can display the image with **matplotlib**:

```
plt.imshow(img)
plt.show()
```

In Python, we represent images as **multi-dimensional arrays**. To access a single **pixel**, we can use the normal array operations:

```
row_index = 10
col_index = 42
img[row_index][col_index]
```

This output shows the **rgb** values of the specified pixel. the **dtype** defines, what data type is used for the values. \\
In this case, **uint8** (unsigned 8-bit integer) spans from 0 to 255 where **0 means black** and **255 means white**.

To convert to another data type, we can use **numpy**:

```
img_float = img.astype(np.float32) / 255
img_float[row_index][col_index]
```

To view the **dimensions** of an image, we can use the **shape attribute** with the format (height, width, channels):

```
img.shape
```

To reduce the **spacial resolution**, we can the **OpenCV** function **resize**.

```
img_small = cv2.resize(img, (50, 50))
plt.imshow(img_small)
plt.show()
```

```
img_small.shape
```

To reduce the **color resolution**, we can use **numpy**. The // operator is an integer division that cuts off the decimal places:

```
img_low_color = img // 64 * 64
plt.imshow(img_low_color)
plt.show()
```
